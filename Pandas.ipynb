UPDATING DATA

```python
#adding new columns
df['defStat']=df['Defense']+df['Sp. Def']
df['atkStat']=df['Attack']+df['Sp. Atk']
df.head(10)
```

```python
#remove columns
df=df.drop(columns=['defStat','atkStat'])
df.head(10)
```

```python
#good way of doing
df['stPoints']=df.iloc[:,4:10].sum(axis=1)
df.head(20)
```

```python
#rearranging columns
cols=list(df.columns)
print(cols)

#stPoints to right of hp
df = df[cols[0:5] + [cols[-1]] + cols[5:12]]
print(df.head(5))
```

DATA FILTERING

```python
#Condtional operations are &,|
df.loc[(df['Type 1']=='Fire') & (df['Type 2']=='Flying') & (df['HP']>70)]
```

```python
#reseting index according to new data frames
new_df=df.loc[(df['Type 1']=='Fire') | (df['Type 2']=='Flying')].head(5).reset_index()
print(new_df)
```

```python
#filtering data based on substrings //names containing mega
new_df=df.loc[(df['Name'].str.contains('Mega'))]
print(new_df)
```

```python
#names not containing mega
new_df=df.loc[~df['Name'].str.contains('Mega')]
print(new_df)
```

```python
#data filtering using reg ex
import re
new_df=df.loc[df['Type 1'].str.contains('Fire|Grass',regex=True)]
print(new_df)
```

```python
#updating df using conditions
df.loc[df['Type 1']=='Fire','Type 1']='Inferno'
print(df.loc[df['Type 1']=='Inferno'])
```

```python
#all legendary pokemons are type 1 inferno
df.loc[df['Type 1']=='Inferno','Legendary']=True
print(df.loc[df['Type 1']=='Inferno'].head(20))
```

STATISTICS

```python
#first select numeric columns then apply mean function
df=pd.read_csv('CSV_FILES\modified')
numeric_cols=df.select_dtypes(include=['int64', 'float64']).columns
grouped_by_type1 = df.groupby('Type 1')[numeric_cols]
mean_values = grouped_by_type1.mean().sort_values('HP',ascending=False)
print(mean_values)

```

```python
#sorting according to type1 pokemons who have the highest atk
mean_atk=grouped_by_type1.mean().sort_values('Speed',ascending=False)
print(mean_atk)
```
