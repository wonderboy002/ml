```python
import matplotlib.pyplot as pl 
import pandas as pd 
import numpy as np 
```

SIMPLE PLOTS

```python
y=np.array([14,-5,56],dtype='int16')
x=np.array([1,2,3],dtype='int16')
pl.plot(x,y)
pl.title('A Simple Plot')#adding title for the graph
#adding labels for axes
pl.xlabel('X-axis')
pl.ylabel('Y-axis')
pl.show()
```

```python
#changing font styles of labels and title using font dict
y=np.array([14,-5,56],dtype='int16')
x=np.array([1,2,3],dtype='int16')
pl.plot(x,y)
pl.title('A Simple Plot',fontdict={'fontname' : 'Comic sans MS','fontsize' : 15})#adding title for the graph
#adding labels for axes
pl.xlabel('X-axis')
pl.ylabel('Y-axis')
pl.show()
```

```python
#adding custom x and y ticks
y=np.array([0,9,18,27,54,66,89],dtype='int16')
x=np.array([0,1,16,32,48,64,80],dtype='int16')
pl.plot(x,y)
pl.title('A Simple Plot',fontdict={'fontname' : 'Comic sans MS','fontsize' : 15})
pl.xticks([0,20,40,60,80])
pl.yticks([0,20,40,60,80])
pl.xlabel('X-axis')
pl.ylabel('Y-axis')
pl.show()
```

```python
#adding legends and other custom params for styling
y=np.array([0,10,20,30,40,50],dtype='int16')
x=np.array([0,20,40,60,80,100],dtype='int16')
pl.plot(x,y,label='HalfY',color='green',linewidth=2,marker='.',linestyle='--',markersize=10,markeredgecolor='purple')#customs params like color,linewidth for thickness,linestyle and marker for data points and markersize+markeredgecolor
#shorthand for above [color][marker][line]
pl.plot(x,y,'b^--',label='HalfY')
pl.title('A Simple Plot',fontdict={'fontname' : 'Comic sans MS','fontsize' : 15})
pl.xticks([0,20,40,60,80,100])
pl.yticks([0,10,20,30,40,50])
pl.xlabel('X-axis')
pl.ylabel('Y-axis')
pl.legend()
pl.show()
```

```python
#adding another plot
y=np.array([0,10,20,30,40,50],dtype='int16')
x=np.array([0,20,40,60,80,100],dtype='int16')
arr=np.arange(0,10.5,0.5)
print(arr)
pl.plot(x,y,label='HalfY',color='green',linewidth=2,marker='.',linestyle='--',markersize=10,markeredgecolor='purple')
pl.plot(arr[:10],arr[:10]**2,color='red')
pl.plot(arr[9:],arr[9:]**2,'b^--',label='squareX',color='black')
pl.legend()
pl.savefig('first.png',dpi=300)
pl.show()
```

###BARCHARTS

```python
labels=['A','B','C','D','E']
values=[100,200,500,50,300]
bars=pl.bar(labels,values,label='bar',color='green')
patterns=['/','o','*','/','o']
for bar in bars:
    bar.set_hatch(patterns.pop())
pl.legend()
pl.show()
```

PLOTTING DATA RELATED TO CSV

```python
gas=pd.read_csv('CSV_FILES\gas_prices.csv')
# print(gas.Year)
#Plotting gas prices of usa on yearly basis
pl.title('Gas Prices over time in $USD',fontdict={'fontname' : 'Arial Black','fontsize' : 15})
pl.plot(gas['Year'],gas['USA'],'b.-',label='USA')
pl.plot(gas['Year'],gas['Japan'],'g.--',label='Japan')
pl.plot(gas['Year'],gas['South Korea'],'r.-',label='South Korea')
pl.plot(gas['Year'],gas['Australia'],'y.-',label='Australia')
pl.legend()
pl.xticks(gas.Year[::3])
pl.xlabel('year')
pl.ylabel('USD')
# print(gas)
pl.savefig('gas_prices.png',dpi=300)
pl.show()
# print(gas)
```

```python
fifa=pd.read_csv('./CSV_FILES/fifa_data.csv')
print(fifa)
```

HISTOGRAMS

```python
pl.hist(fifa['Age'])
pl.show()
```

```python
#plotting histogram on number of players having age within a range
bins=np.arange(15,46,5)
print(bins)
pl.hist(fifa['Age'],bins=bins,color='#00b894')
pl.xlabel('Age Groups')
pl.ylabel('Number of players')
pl.title('Distibution of age groups of players in fifa')
pl.show()
```

PIE CHARTS

```python
#a simple pie charts
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()
# print(left.iloc[0])
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()
# print(right.iloc[0])
labels=['Left Footers','Right Footers']
colors=['#ff4757','#3742fa']
pl.title('Foot Preference of fifa players')
pl.pie([left.iloc[0],right.iloc[0]],labels=labels,colors=colors,autopct='%.2f %%')
#customization can be done for labels,colors,and decimal points
pl.show()
```

```python
fifa['Weight']=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
print(fifa['Weight'][0])
```

```python
#pie chart based on distribution of player's weights in fifa
light_mid=fifa.loc[(fifa['Weight']>=125) & (fifa['Weight']<150)].count().iloc[0]

heavy_light=fifa.loc[(fifa['Weight']>=150) & (fifa['Weight']<175)].count().iloc[0]
heavy_mid=fifa.loc[(fifa['Weight']>=175) & (fifa['Weight']<200)].count().iloc[0]
heavy=fifa.loc[fifa['Weight']>200].count().iloc[0]
weights=np.array([light_mid,heavy_light,heavy_mid,heavy],dtype='int16')
labels=['Light Medium','Heavy Light','Heavy Mid','Heavy']
colors=['#00b894','#d63031','#6c5ce7','#0984e3']
pl.pie(weights,labels=labels,colors=colors,autopct='%.2f %%')
pl.show()
```
